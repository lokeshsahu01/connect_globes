```python
counter = 0
PN = []
TI = []
VO = []
LA = []
DA = []
IM = []
CR = []
bigcount =18000000 
filecount = 126
while bigcount < 18700000:
    endpoint = bigcount + 100001
    print('starting point is ', bigcount)
    print('the end point is' , endpoint)
    with open('D:/GB_DB/inventory.bson','rb') as f:
        data = bson.decode_file_iter(f)
        data2 = itertools.islice(data,bigcount,endpoint)
        print((data2))
        for line in data2:
            if counter < 100000:
    #             print(line)
    #             print((line))          
    #             print(line['PERSONALNO'])
    #             print(line['TITLE'])
    #             print(line['VOLUME'])
    #             print(line['CREATOR'])
    #             print(line['LANGUAGE'])
    #             print(line['DATE'])
    #             print(line['IMAGECOUNT'])

                PN.append(line['PERSONALNO'])
                TI.append(line['TITLE'])
                LA.append(line['LANGUAGE'])
                st = (line['DATE'])
    #             print('this is str', line['DATE'])
                a = ''
                for i in str(st):
                    if i.isdigit():
                        a = a+i
    #             print(a)
                DA.append(a)
                IM.append(line['IMAGECOUNT'])
    #             try:

    #                 VO.append(line['VOLUME'])
    #             except KeyError:
    #                 line['VOLUME'] = 'Empty'
    #                 VO.append(line['VOLUME'])
                CR.append(line['CREATOR'])

            else:
                break
            counter = counter + 1

    bigcount = bigcount + counter
    filecount = filecount + 1
    df = pd.DataFrame({'Personal Number':PN, 'title' : TI, 'Lang': LA, 'Date': DA, 'Creator': CR, 'Pages': IM  })
    filename = str(filecount) + 'file.csv'
    df.to_csv(filename)
    
    print(counter)
    print('Code is processed')
    print('the files traversed are', bigcount)
    counter = 0
    
    
    #from here the CSV is being read
    data = pd.read_csv(filename, encoding="ISO-8859-1")
    print(len(data))
    len(data['Personal Number'])
    df1 = data.replace(np.nan, 'Empty', regex=True)

    df1['newtitle'] = df1['Personal Number'].astype(str) + df1['title']
    df1['newdate'] = df1['Personal Number'].astype(str) + df1['Date'].astype(str)
    df1['newcreator'] = df1['Personal Number'].astype(str) + df1['Creator']
#     print(df1['newcreator'][4])
    gc.collect()
    #finding first match
    matches = match_strings(df1['newtitle'])
    #finding second match
    matches2 = match_strings(df1['newcreator'])
    #finding third match
    matches3 = match_strings(df1['newdate'])



    new1 = (matches[matches.left_side != matches.right_side])
    new1.head()
    new1['PN'] = new1['left_side'].str[:13]
    new1['left_side'] = new1['left_side'].str[13:]
    new1['PN_right'] = new1['right_side'].str[:13]
    new1['right_side'] = new1['right_side'].str[13:]
    titlefile = str(filecount) + 'titleduplicacy.csv'
    new1.to_csv(titlefile)
    print('lot one done')
    newcreator = (matches2[matches2.left_side != matches2.right_side])
    # new1.head()
    newcreator['PN'] = newcreator['left_side'].str[:13]
    newcreator['left_side'] = newcreator['left_side'].str[13:]
    newcreator['PN_right'] = newcreator['right_side'].str[:13]
    newcreator['right_side'] = newcreator['right_side'].str[13:]
    creatorfile = str(filecount) + 'creatorduplicacy.csv'
    newcreator.to_csv(creatorfile)
    print('second matching done')
    newdate = (matches3[matches3.left_side != matches3.right_side])
    print('third matching done')
    # new1.head()
    newdate['PN'] = newdate['left_side'].str[:13]
    newdate['left_side'] = newdate['left_side'].str[13:]
    newdate['PN_right'] = newdate['right_side'].str[:13]
    newdate['right_side'] = newdate['right_side'].str[13:]
    datefile = str(filecount) + 'dateduplicacy.csv'
    newcreator.to_csv(datefile)
    gc.collect()
    PN = []
    TI = []
    VO = []
    LA = []
    DA = []
    IM = []
    CR = []
    
    

```
